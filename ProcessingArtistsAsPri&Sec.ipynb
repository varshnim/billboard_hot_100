```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/MusicDataProcessed.csv")
```

```python
data['Song'] = data['Song'].replace(['"'],'',regex=True)
data['Artist(s)'] = data['Artist(s)'].replace('featuring',',',regex=True)
data['pri_artist'] = [i.split(",")[0] for i in data['Artist(s)']]
data['sec_artist(s)'] = [i.split(",")[1] if len(i.split(","))>=2 else "None" for i in data['Artist(s)']]
data['sec_artist(s)'] = data['sec_artist(s)'].str.replace(" and ",",",regex=True)
data['sec_artist(s)'] = [i.strip(" ") for i in data['sec_artist(s)']]
data['sec_art_1'] = [i.split(",")[0] if i!="None" else "None" for i in data['sec_artist(s)']]
data['sec_art_2'] = [i.split(",")[1] if i!="None" and len(i.split(","))>=2 else "None" for i in data['sec_artist(s)']]
data.drop(['sec_artist(s)'], axis=1, inplace=True)
```
