```python
import pandas as pd
import requests
import numpy as np
```

```python
url_body = "https://en.wikipedia.org/wiki/List_of_Billboard_Hot_100_number_ones_of_"
```

```python
bbhh_dict = {}
```

```python
for i in range(1971,2021,1):
  page = pd.read_html(url_body+str(i))
  for content in page:
    if len(content)==51 or len(content)==52 or len(content)==53:
      temp = content
      break
  temp['year']=i
  bbhh_dict[i]=temp
```

```python
bbhh_df = pd.concat(bbhh_dict.values())
```

```python
bbhh_df['Issue date']=bbhh_df['Issue date'].replace(np.nan,'_')
bbhh_df['Issue date[1]'] = bbhh_df['Issue date[1]'].replace(np.nan,'_')
```

```python
bbhh_df['Date'] = bbhh_df['Issue date']+bbhh_df['Issue date[1]']
bbhh_df['Date'] = [i.strip("_") for i in bbhh_df['Date']]
bbhh_df['Date'] = bbhh_df['Date'].replace("\xa0"," ",regex=True)
```

```python
bbhh_df.drop(['Issue date','Issue date[1]'], axis=1, inplace=True)
```

```python
# bbhh_df.to_csv("scrapped_raw_1971-2020.csv",index=False)
```

```python
bbhh_df['Month'] = [i.split(" ")[0] for i in bbhh_df['Date']]
bbhh_df['Day'] = [i.split(" ")[1] for i in bbhh_df['Date']]
```

```python
bbhh_df.drop(['Date'], axis=1, inplace=True)
```

```python
month_dict ={
    'january':1,
    'february':2,
    'march':3,
    'april':4,
    'may':5,
    'june':6,
    'july':7,
    'august':8,
    'september':9,
    'october':10,
    'november':11,
    'december':12
}
```

```python
bbhh_df['Month'] = [i.lower() for i in bbhh_df['Month']]
bbhh_df['Month']=bbhh_df['Month'].map(month_dict)
```

```python
bbhh_df.drop(['No.','Reference','Ref.','Reference(s)'], axis=1, inplace=True)
```
