```python
import pandas as pd
```

```python
data70 = pd.read_csv("/content/lyrics_70s.csv", header=None).T
data70.columns = ['artist*song','lyrics']
```

```python
data80 = pd.read_csv("/content/lyrics_80s.csv", header=None).T
data80.columns = ['artist*song','lyrics']
```

```python
data90 = pd.read_csv("/content/lyrics_90s.csv", header=None).T
data90.columns = ['artist*song','lyrics']
```

```python
data2000 = pd.read_csv("/content/lyrics_2000s.csv", header=None).T
data2000.columns = ['artist*song','lyrics']
```

```python
data2010 = pd.read_csv("/content/lyrics_2010s.csv", header=None).T
data2010.columns = ['artist*song','lyrics']
```

```python
basedata = pd.read_csv("/content/bbhh_lyrics_url.csv")
basedata
```

```python
seventies = list(range(1971,1981,1))
eighties = list(range(1981,1991,1))
nineties = list(range(1991,2001,1))
twothous = list(range(2001,2011,1))
twothoutens = list(range(2011,2021,1))
```

```python
bd70 = basedata[basedata['year'].isin(seventies)]
bd80 = basedata[basedata['year'].isin(eighties)]
bd90 = basedata[basedata['year'].isin(nineties)]
bd2000 = basedata[basedata['year'].isin(twothous)]
bd2010 = basedata[basedata['year'].isin(twothoutens)]
```

```python
bd70['artist*song'] = bd70['Song']+'*'+bd70['pri_artist']
bd80['artist*song'] = bd80['Song']+'*'+bd80['pri_artist']
bd90['artist*song'] = bd90['Song']+'*'+bd90['pri_artist']
bd2000['artist*song'] = bd2000['Song']+'*'+bd2000['pri_artist']
bd2010['artist*song'] = bd2010['Song']+'*'+bd2010['pri_artist']
```

```python
fdata70 = bd70.merge(data70, on='artist*song', how='left')
fdata80 = bd80.merge(data80, on='artist*song', how='left')
fdata90 = bd90.merge(data90, on='artist*song', how='left')
fdata2000 = bd2000.merge(data2000, on='artist*song', how='left')
fdata2010 = bd2010.merge(data2010, on='artist*song', how='left')
```

```python
fulldata = pd.concat([fdata70,fdata80,fdata90,fdata2000,fdata2010], axis=0)
```
