# Dictionary to Clean Song Titles & Primary Artists 

```python
splc_replacement = {
    '&':'and',
    'à':'a',
    'è':'e',
    'ì':'i',
    'ò':'o',
    'ù':'u',
    'á':'a',
    'é':'e',
    'í':'i',
    'ó':'o',
    'ú':'u',
    'ý':'y',
    'â':'a',
    'ê':'e',
    'î':'i',
    'ô':'o',
    'û':'u',
    'ã':'a',
    'ñ':'n',
    'õ':'o', 
    "\.":"",
    "\,":"",
    "\'":"",
    "\(":"",
    "\)":"",
    "\[":"",
    "\]":"",
    "\”":"",
    "\–":" ",
    "\\/":"",
    "\!":"",
}
```

```python
data = pd.read_csv("/content/BBHH-Processed.csv")
```

```python
urld = data
```

# Editing song titles for URL

```python
urld['SongTitle'] = urld['Song']
urld['Song'] = urld['Song'].str.lower()
urld['Song'] = urld['Song'].str.strip(" ")
urld['Song'] = urld['Song'].str.replace("\[\w\]","", regex=True)
urld['Song'] = urld['Song'].replace(splc_replacement, regex=True)
# urld['Song'] = urld['Song'].str.replace("&","and")
# urld['Song'] = urld['Song'].str.replace("í","i")
# urld['Song'] = urld['Song'].str.replace("ñ","n")
# urld['Song'] = urld['Song'].str.replace("\,|\'|\(|\)|\”|\'|\.|\[|\]|","")
urld['Song'] = urld['Song'].str.replace("/| |–"," ")
# urld['Song'] = urld['Song'].str.replace("---|--","-")
```

# Editing Primary Artist's Name for URL

```python
urld['PriArtName'] = urld['pri_artist']
urld['pri_artist'] = urld['pri_artist'].str.strip(" ")
urld['pri_artist'] = urld['pri_artist'].str.replace("\[\w\]","", regex=True)
urld['pri_artist'] = urld['pri_artist'].replace(splc_replacement, regex=True)
urld['pri_artist'] = urld['pri_artist'].str.lower()
# urld['pri_artist'] = urld['pri_artist'].str.replace("\,|\'|\(|\)|\”|\'|\.|\[|\]|\/|","")
# urld['pri_artist'] = urld['pri_artist'].str.replace("&","and")
# urld['pri_artist'] = urld['pri_artist'].str.replace(" ","-")
# urld['pri_artist'] = urld['pri_artist'].str.replace("é","e")
```
